```python
# Importar librerías
import earthpy.plot as ep
import numpy as np
import matplotlib.pyplot as plt
import rasterio 
```

```python
# Cargar archivos desde repositorios
url = 'src/Sentinel/gdal/Sentinel_10m.tif'
url2 = 'src/Sentinel/gdal/Sentinel_20m.tif'
archivo = rasterio.open(url)
```

### Visualización composición de color

```python
#
# Definición de variables para presentación de imagen
#
rgb = (3,2,0)
figsize = (20,10)
title = 'Imágen área de estudio composición {}'.format(rgb) 
#
# Presentación de imagen con composición específica
#
ep.plot_rgb(archivo.read(), rgb=rgb, title=title, stretch=True, figsize=figsize )
```

### Visualización por bandas

```python
#
# Presentación de imagen por bandas
#
titles=["Banda 1", "Banda 2", "Banda 3", "Banda 4"]
figsize = (15,15)
ep.plot_bands(archivo.read(), title=titles, figsize=figsize, cbar=False)
plt.show()
```

```python
# Histogramas de las bandas
ep.hist(archivo.read(), title=titles)
plt.show()
```
